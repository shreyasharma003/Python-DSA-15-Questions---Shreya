#Question02: Coin Change Problem

```python
from typing import List

def coin_change(coins: List[int], amount: int) -> int:

    dp = [amount + 1] * (amount + 1)

    dp[0] = 0   

    for i in range(1, amount + 1):
        for coin in coins:
            if i - coin >= 0:
                dp[i] = min(dp[i], 1 + dp[i - coin])

    min_coins = dp[amount] if dp[amount] != amount + 1 else -1

    return min_coins


print(f"Input: coins=[1, 2, 5], amount=11, Output: {coin_change([1, 2, 5], 11)}")
print(f"Input: coins=[2], amount=3, Output: {coin_change([2], 3)}")
```

#Question03: Find Zero Sum Triplets

```python
def find_zero_sum_triplets(nums: list[int]) -> list[list[int]]:
    nums.sort()                
    result = []
    n = len(nums)

    for i in range(n):

        if i > 0 and nums[i] == nums[i - 1]:
            continue

        a = nums[i]
        left = i + 1
        right = n - 1


        while left < right:
            total = a + nums[left] + nums[right]

            if total == 0:
                triplet = [a, nums[left], nums[right]]
                result.append(triplet)

              
                left += 1
                while left < right and nums[left] == nums[left - 1]:
                    left += 1

          
                right -= 1
                while left < right and nums[right] == nums[right + 1]:
                    right -= 1

            elif total < 0:
                left += 1
            else:
                right -= 1

    return result

print(f"Input: [-1, 0, 1, 2, -1, -4], Output: {find_zero_sum_triplets([-1, 0, 1, 2, -1, -4])}")
print(f"Input: [0, 1, 1], Output: {find_zero_sum_triplets([0, 1, 1])}")
```

#Question04:Sum Array Elements if length is Prime

```python
def sum_if_length_prime(arr: list[int]) -> int:

    def is_prime(n):
        if n <= 1:
            return False
        if n == 2:
            return True
        if n % 2 == 0:
            return False
        for i in range(3, int(n**0.5) + 1, 2):
            if n % i == 0:
                return False
        return True

    length = len(arr)

    if is_prime(length):
        return sum(arr)
    else:
        return 0


print(f"Input: [1, 2, 3], Output: {sum_if_length_prime([1, 2, 3])}") 
print(f"Input: [10, 20, 30, 40], Output: {sum_if_length_prime([10, 20, 30, 40])}")
```

#Question09: Two Sum

```python
def two_sum(nums: list[int], target: int) -> list[int]:
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []

print(two_sum([2, 7, 11, 15], 9))
```

#Question08 : Valid Parantheses

```python
def is_valid(s: str) -> bool:
   
    stack = []
    
   
    pairs = {
        ')': '(',
        '}': '{',
        ']': '['
    }

    for char in s:
        
        if char in "({[":
            stack.append(char)
        else:
            
            if not stack:
                return False
        
            
            top = stack.pop()

          
            if pairs[char] != top:
                return False

   
    is_string_valid = len(stack) == 0
  

    return is_string_valid


print(f"Input: \"()\", Output: {is_valid('()')} ")
print(f"Input: \"()[]{{}}\", Output: {is_valid('()[]{}')} ")
```

#Question 7: Increment Large Integer Represented as Digits

```python
def increment_digits(digits: list[int]) -> list[int]:
    
    for i in range(len(digits) - 1, -1, -1):
        if digits[i] < 9:
            digits[i] += 1
            return digits
        else:
            digits[i] = 0
    return [1] + digits


# Example calls for testing
print(f"Input: [1,2,3], Output: {increment_digits([1,2,3])}")
print(f"Input: [9,9], Output: {increment_digits([9,9])}")
```

#Question 11: Transpose a Matrix

```python
def transpose_matrix(matrix: list[list[int | float]]) -> list[list[int | float]]:

    if not matrix or not matrix[0]:
        return []
        
    rows = len(matrix)
    cols = len(matrix[0])
    
    transposed_result = []
    
    for c in range(cols):
        new_row = []

        for r in range(rows):
            new_row.append(matrix[r][c])
        transposed_result.append(new_row)


    return transposed_result

# Example Test Cases (you can add more to test your code)
print(f'Transpose of [[1, 2], [3, 4]] = {transpose_matrix([[1, 2], [3, 4]])}')
print(f'Transpose of [[1, 2, 3], [4, 5, 6]] = {transpose_matrix([[1, 2, 3], [4, 5, 6]])}')

```

#Question 13: Find the Single Unique Number

```python
def find_single_number(nums: list[int]) -> int:
    single_number = 0
    
    for n in nums:
        single_number ^= n  
   
    return single_number

print(f"Input: [2, 2, 1], Output: {find_single_number([2, 2, 1])}")
print(f"Input: [4, 1, 2, 1, 2], Output: {find_single_number([4, 1, 2, 1, 2])}")
```

#Question 12: Matrix Multiplication

```python
def multiply_matrices(matrix_a: list[list[float]], matrix_b: list[list[float]]) -> list[list[float]]:
  
    rows_a = len(matrix_a)
    cols_a = len(matrix_a[0])
    rows_b = len(matrix_b)
    cols_b = len(matrix_b[0])


    if cols_a != rows_b:
        return []

    product_matrix = [[0 for _ in range(cols_b)] for _ in range(rows_a)]


    for i in range(rows_a):
        for j in range(cols_b):
        
            for k in range(cols_a):
                product_matrix[i][j] += matrix_a[i][k] * matrix_b[k][j]

    return product_matrix



matrix_a_1 = [[1.0, 2.0], [3.0, 4.0]]
matrix_b_1 = [[5.0, 6.0], [7.0, 8.0]]
print(f"Input matrix_a: {matrix_a_1}, matrix_b: {matrix_b_1}, Output: {multiply_matrices(matrix_a_1, matrix_b_1)}")

# Example 2: Incompatible dimensions
matrix_a_2 = [[1.0, 2.0, 3.0]]
matrix_b_2 = [[4.0], [5.0]]
print(f"Input matrix_a: {matrix_a_2}, matrix_b: {matrix_b_2}, Output: {multiply_matrices(matrix_a_2, matrix_b_2)}")
```
